```python
#continue: Goes to the top of the closest enclosing loop

mystring='Sammy'
```

```python
for letter in mystring:
    if letter=='a':
        continue #goes back to the closest enclosing loop(not if!! no statement!!), which is >for letter~mystring:<
    print(letter)
```

```python
#break: Breaks out of the current closest enclosing loop

for letter in mystring:
    if letter=='a':
        break #stop this loop if that letter is equal to 'a'
    print(letter)
```
